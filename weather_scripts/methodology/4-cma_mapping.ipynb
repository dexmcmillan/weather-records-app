```python
import geopandas
```

```python
cmas = geopandas.read_file("https://www12.statcan.gc.ca/census-recensement/2021/geo/sip-pis/boundary-limites/files-fichiers/lcma000b21a_e.zip").to_crs("EPSG:4326")

cmas
```

```python
data = cmas.copy()

data["geometry"] = data["geometry"].centroid
data["Latitude"] = data["geometry"].x
data["Longitude"] = data["geometry"].y
data = data[["CMANAME", "Latitude", "Longitude"]].set_index("CMANAME")

data
```

```python
data.to_csv("./cma_geodata.csv")
```
